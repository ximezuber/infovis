```python
import pandas as pd
import altair as alt
from vega_datasets import data
```

# MakeoverMonday: Week 22 - The Plastic Waste Makers Index

## Import del dataset

```python
df = pd.read_excel('./plasticWasteDataset.xlsx')
df_red = df.loc[df['Rank'] < 31]
df_red
```

## Bar Chart de los 30 productores con más desperdicio de plástico 

```python
alt.Chart(df_red).transform_fold(
    ['Flexible format contribution to SUP waste', 'Rigid format contribution to SUP waste'],
    as_=['Format', 'value']
).mark_bar(
    cornerRadiusBottomRight=3,
    cornerRadiusTopRight=3
).encode(
    x='value:Q',
    y=alt.Y('Polymer Producer', sort='-x'),
    color='Format:N'
).configure_range(
    category={'scheme': 'set2'}
)
```

## Scatter Plot del desperdicio contra la cantidad de activos de cada productor

```python
alt.Chart(df).mark_point().encode(
    x='Assets',
    y='Total contribution to SUP waste',
    color=alt.Color('Rank', scale=alt.
                      Scale(scheme = 'darkred')),
    tooltip=['Polymer Producer', 'Rank', 'Total contribution to SUP waste']
).interactive()
```
